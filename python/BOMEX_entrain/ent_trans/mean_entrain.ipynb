```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import glob
%matplotlib inline
```

```python
plt.style.use('ggplot')
file = Dataset('BOMEX_CORE_ENTRAIN_256x256x128_25m_25m_1s_16_0000010860.nc')

z_vals = file['z']
q_entrain = file['EQTETCOR'][0]
hor_mean = np.mean(q_entrain, axis=(1,2))

fig,ax = plt.subplots(1,1,figsize=[12,12])
ax.plot(hor_mean,z_vals,color='green')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
out = ax.set(title='Horizontal Mean Specific Humidity Entrainment from BOMEX (First Timestep)',
             xlabel='Horizontal Mean Entrainment (kg m^-3 s^-1)',
             ylabel='Height (m)', xlim=[-2.5e-7,4e-6],ylim=[0,2500])
```
